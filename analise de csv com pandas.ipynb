```python
import pandas as pd

doc = pd.read_csv('dataset.csv')
```

```python
doc.head(5)
```

```python
doc.isna().sum()
```

```python
#calculando a moda
#MODA: em estatistica moda é uma medida de tendencia central que representa
#o valor mais frequente de um conjunto de dados
#moda é especialmente útil quando queremos saber qual é o valor mais comum
#ou popular em um conjunto de dados 
```

```python
#moda
moda = doc['Quantidade'].value_counts().index[0]
```

```python
print(moda)
```

```python
#a função fillna() é usada para preencher os valores ausentes. A função oferece
#muitas opções. Podemos usar um valor específico, uma função agregada(como media)
#ou o valor anterior ou seguinte
```

```python
#preenchendo valores ausentes como a moda 

doc['Quantidade'].fillna(value=moda, inplace = True)
```

```python
doc.isna().sum()
```
