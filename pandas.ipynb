```python
from pandas import DataFrame
from numpy import arange
dados ={'Estado':['Santa Catarina','Rio de Janeiro','Tocantins','Baiha','Minas Gerais'],
       'Ano':[2004,2005,2006,2007,2008],
       'Taxa de Desemprego':[1.5,1.7,1.6,2.4,2.7]}
df = DataFrame(dados)#converte o dicionario em um dataframe
```
