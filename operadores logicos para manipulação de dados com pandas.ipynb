```python
import pandas as pd 

dados_csv = pd.read_csv('dataset.csv')
```

```python
dados_csv.head()
```

```python
#o operados logico and checa as duas condições. Serão retornados os resultados quando as duas condições forem verdadeiras 
dados_csv[(dados_csv.Segmento == 'Home Office') & (dados_csv.Regiao == 'South')].head()
```

```python
#ele retorna se uma das condições forem verdadeiras 
dados_csv[(dados_csv.Segmento == 'Home Office') | (dados_csv.Regiao == 'South')].tail()#tail traz os ultimos registros
```

```python
#filtrando as vendas que não ocorrem no segmento home offoce e south
dados_csv[(dados_csv.Segmento != 'Home Office') & (dados_csv.Regiao != 'South')].sample(5)#sample traz os registros de maneira aleatoria 
```

```python
#agrupamento de dados em data frame

dados_csv[['Segmento','Regiao','Valor_Venda']].groupby(['Segmento','Regiao']).mean()
#mean é media
```

```python
#Agregação multipla 
#Unindo a função groupby() com a função agg() para realizar agrupamento multiplo

dados_csv[['Segmento','Regiao','Valor_Venda']].groupby(['Segmento','Regiao']).agg(['mean','std','count'])
```

No codigo acima primeiro filtramos os dados em tres colunas [Segmento,Regiao,Valor_Venda]. Na sequência, agrupamos por duas colunas [Segmento, Regiao]. E então agregamos os dados calculando média, desvio padrão e contagem de elementos para a coluna que ficou fora do groupby, nesse caso foi Valor_Venda

A função agg() recebe como argumento uma lista de funções para agregação
