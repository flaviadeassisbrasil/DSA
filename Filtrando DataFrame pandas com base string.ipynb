```python
#Filtrando o dataframe pela coluna segmento com valores que iniciam com as letras 'Con'

doc[doc.Segmento.str.startswith('Con')].head()
```

```python
#Filtrando dataframe pela coluna Segmento com valores que terminam com as letras 'mer'

doc[doc.Segmento.str.endswith('mer')].head()
```

```python
#Split de string com data frame 

doc['ID_Pedido'].head() #formato dos dados de id pedido 
```

```python
#split coluna pelo caracter ['-']

doc['ID_Pedido'].str.split('-')
```

O resultado da celula acima mostra que foi divido em várias listas para extrair o ano precisamos especifcar a posição do indice da informação que vai ser extraida, no caso o ano

```python
doc['ID_Pedido'].str.split('-').str[1].head()
```

```python
#criando uma nova coluna com a informação extraida
doc['Ano'] = doc['ID_Pedido'].str.split('-').str[1]
```

```python
doc.head()
```

```python
#STRIP que remove caracteres da string

doc['Data_Pedido']

```

```python
doc['Data_Pedido'] = doc['Data_Pedido'].str.strip('20') 
```

```python
doc.head()
```

```python
#SUBSTINTUINDO CARACTERES 

doc['ID_Cliente'] = doc['ID_Cliente'].str.replace('CG','AX')
```

```python
doc.head()
```

```python
#COMBINAÇÃO DE STRING

doc['Pedido_Segmento'] = doc['ID_Pedido'].str.cat(doc['Segmento'], sep='-')
doc.head()
```
